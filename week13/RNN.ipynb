# RNN 이란?

Recurrent Neural Network, RNN은 Sequence Data를 다루는 대표적인 인공 신경망이다. 

Sequence Data란 순서형 데이터로 주가, 문장, 신호 등의 순서에 의미가 있는 데이터를 말한다.

Sequence Data는 아래의 그림과 같이 다양한 Task를 다룰 수 있는데 이러한 Task를 다루는데 적합한 모델이 바로 RNN 이다.

- one to one: RNN이 아닌 기본적인 인공신경망 모델이 다루는 Task

- one to many: 한장의 이미지를 여러 문장으로 해석하는 모델

- many to one: 문장을 해석해서 하나의 정보로 축약하는 모델(주가 예측의 경우도 여기에 해당한다.)

- many to many 1: 문장의 의미를 분석해서 하나의 context vector를 만들고 다른 문장으로 출력하는 모델(기계 번역)

- many to many 2: 동영상과 같은 실시간 연속 프레임에 대한 분석 결과를 출력하는 모델

RNN은 아래의 그림 처럼 순환 구조를 이루고 았는 인공 신경망으로 Hidden feature를 만들기 위해서 현재 state의 input과 이전 state의 output을 사용한다.

우리가 앞에서 배웠던 신경망들은 one to one Task에서 처럼 input을 은닉층과 활성화 함수를 거쳐서 오직 출력층 방향으로만 향했다. 이러한 신경망을 Feed Forward Neural Network라고 한다. 하지만 RNN의 경우 Feed Forward Neural Network과 달리 input이 출력층의 방향으로만 향하지 않는다.

RNN은 은닉층의 노드에서 활성화 함수를 통해서 나온 결과 값이 출력층 방향으로도 보내면서, 다시 은닉층 노드의 다음 계산의 입력으로 보내는 특징을 갖고 있다. 이러한 특징 덕분에 Sequence Data를 다루기에 용이하다.

image.png

RNN은 위의 그림 처럼 A라는 Cell이 활성화 함수를 통해 내보내는 값이 다음 노드에도 영향을 주기 때문에 과거의 기억을 보존하며, 과거의 데이터에 영향을 받으며 출력 값을 내보낼 수 있게 된다. 여기서 Cell은 일종의 메모리 역할을 하기 때문에 메모리 셀이라고도 표한한다. 메모리 셀이 출력층 방향으로 또는 다음 시점 t+1의 자신에게 보내는 값을 은닉 상태(hidden state)라고 한다. 

종합하자면 RNN은 과거의 기억을 보존한 상태로 hidden feature를 출력하기 위해서 t 시점의 메모리 셀은 t-1 시점의 메모리 셀이 보낸 hidden state값을 t 시점의 hidden state 계산을 위한 입력값으로 사용하게 된다. 그리고 최종 출력 값이 hidden feature이 되며 이 값을 다시 FC Layer를 거쳐서 원하는 값을 출력할 수 있게 된다.

지금 배운 RNN을 가장 단순한 형태의 RNN이라고 하여 Vanilla RNN이라고 불린다. 그런데 이 바닐라 RNN은 비교적 짧은 시퀀스에 대해서만 효과를 보이는 단점이 존재한다. 바닐란 RNN의 시점이 길어질 수록 앞의 정보가 뒤로 충분히 전달되지 못하는 현상이 발생하며 이러한 문제를 장기 의존성 문제(the problem of Long-Term Dependencies)라고 한다. 이러한 장기 의존성 문제를 해결하기 위해서 다양한 RNN의 변형 모델이 나왔고 앞으로 다룰 LSTM, GRU 모두 그 변형 모델들에 속한다.

# LSTM 이란?

Long Short-Term Memory, LSTM은 전통적인 RNN의 단점을 보완한 모델이다. LSTM은 은닉층의 메모리 셀에 입력 게이트, 망각 게이트, 출력 게이트를 추가하여 불필요한 기억을 지우고, 기억해야할 것들을 정한다.

따라서 LSTM은 은닉 상태(hidden state)를 계산하는 식이 전통적인 RNN보다 조금 더 복잡해졌으며 셀 상태(cell state)라는 값이 추가되었다.

덕분에 LSTM은 RNN과 비교하여 긴 시퀀스의 입력을 처리하는데 탁월한 성능을 보여준다.

# GRU 이란?

Gate Recurrent Unit, GRU는 LSTM의 gate를 줄인 모델이다. LSTM은 gate의 수가 너무 많아 구조적으로 복잡하기 때문에 연산속도 면에서 많은 손해를 입게 된다. 따라서 RNN의 단점을 보완하면서 LSTM의 효율성을 높이기 위한 모델에 대해서 연구가 시작되었고 그 연구의 결과가 바로 GRU이다.

GRU는 cell state 없고 hidden state만이 존재하며 망각 게이트와 출력 게이트를 결합시키고 초기화 게이트가 추가된 형태로 LSTM보다 조금 더 효율성이 높은 모델이라고 할 수 있다.

RNN의 단점을 보완하고자 LSTM, GRU 등의 모델이 개발되었음 에도 여전히 RNN의 vanishing gradient 문제는 존재했다.
이러한 RNN의 vanishing gradient 문제를 해결하고자 다른 접근법이 연구되었고 그 결과로 Attention 기법이 개발되었다. 그리고 이 Attention 기법만을 활용한 새로운 모델이 바로 Transformer이다. 앞으로 RNN에 대하여 더욱더 공부하고자 한다면 Attention과 Transformer을 공부하는 것이 도움이 될 것이다.

# RNN실습

대표적인 시게열 데이터인 주가 데이터를 가지고 LSTM 모델을 통해서 주가를 예측하는 실습을 진행했다.

데이터 출처: https://finance.yahoo.com/quote/VTI/history?p=VTI

```python
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from copy import deepcopy

from datetime import date
from numpy.random import seed
from pylab import rcParams
from sklearn.metrics import mean_squared_error
from tqdm import tqdm_notebook
from sklearn.preprocessing import StandardScaler

import torch
import torch.nn as nn
import torch.nn.functional as F
from copy import deepcopy

import warnings

warnings.filterwarnings("ignore")
```

```python
if torch.cuda.is_available():
  DEVICE = torch.device('cuda')
else:
  DEVICE = torch.device('cpu')
```

```python
test_size = 0.2

seq_length = 5 # 들어가는 데이터의 수
data_dim = 6 # 컬럼의 수
hidden_dim = 10 # 저장되는 값 
output_dim = 1 # 맞추고자 하는 값

lr = 0.01
EPOCHS = 500
BATCH_SIZE = 64
```

```python
PATHS = '/content/drive/MyDrive/PD학기제/Week13, RNN/'

df = pd.read_csv(PATHS + 'VTI.csv')

df.loc[:, 'Date'] = pd.to_datetime(df['Date'],format='%Y-%m-%d')

df.columns = [str(x).lower().replace(' ', '_') for x in df.columns]

df['month'] = df['date'].dt.month

df.sort_values(by='date', inplace=True, ascending=True)

df.head()
```

```python
ax = df.plot(x='date', y='adj_close', style='b-', grid=True)
ax.set_xlabel("date")
ax.set_ylabel("USD")
```

```python
num_test = int(test_size*len(df))
num_train = len(df) - num_test

print("num_train = " + str(num_train))
print("num_test = " + str(num_test))

train_df = df[:num_train]
test_df = df[num_train:]

print("train.shape = " + str(train_df.shape))
print("test.shape = " + str(test_df.shape))
```

```python
ax = train_df.plot(x='date', y='adj_close', style='b-', grid=True)
ax = test_df.plot(x='date', y='adj_close', style='g-', grid=True, ax=ax)
ax.legend(['train', 'test'])
ax.set_xlabel("date")
ax.set_ylabel("USD")
```

```python
# scaling function for input data
def build_dataset(time_series, seq_length):
    dataX = []
    dataY = []
    for i in range(0, len(time_series) - seq_length):
        _x = time_series[i:i + seq_length, :]
        _y = time_series[i + seq_length, [-2]]  # Next close price
        dataX.append(_x)
        dataY.append(_y)
    return np.array(dataX), np.array(dataY)
```

```python
scaler = StandardScaler()
scaler.fit(df.drop(labels=['date', 'month'], axis=1))
train = scaler.transform(train_df.drop(labels=['date', 'month'], axis=1))
train_x, train_y = build_dataset(time_series = train, seq_length = seq_length)

test = scaler.transform(test_df.drop(labels=['date', 'month'], axis=1))
test_x, test_y = build_dataset(time_series = test, seq_length = seq_length)

print(f"train_x.shape = {train_x.shape}, train_y.shape = {train_y.shape}")
print(f"test_x.shape = {test_x.shape}, test_y.shape = {test_y.shape}")
```

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

class CustomDataset(Dataset):

  # 데이터 정의
  def __init__(self, x_data, y_data = None):
    self.x_data = x_data
    self.y_data = y_data

  # 이 데이터 셋의 총 데이터 수
  def __len__(self):
    return len(self.x_data)

  # 어떠한 idx를 받았을 때 그에 맞는 데이터를 반환
  def __getitem__(self, idx):
    if self.y_data is None:
      x = torch.Tensor(self.x_data[idx])
      return x
    else:
      x = torch.Tensor(self.x_data[idx])
      y = torch.Tensor(self.y_data[idx])
      return x, y
```

```python
def train(model, train_loader):
  model.train()
  train_loss = 0

  for batch_idx, (feature, target) in enumerate(train_loader):

    feature = feature.to(DEVICE)
    target = target.to(DEVICE)
    optimizer.zero_grad()
    outputs = model(feature)
    loss = criterion(outputs, target)
    loss.backward()
    optimizer.step()
    train_loss += loss.item()

  train_loss /= len(train_loader)
  return train_loss

def evaluate(model, test_loader):
  model.eval()
  test_loss = 0

  with torch.no_grad():
    for batch_idx, (feature, target) in enumerate(test_loader):
      feature = feature.to(DEVICE)
      target = target.to(DEVICE)
      outputs = model(feature)
      loss = criterion(outputs, target)
      test_loss += loss.item()

  test_loss /= len(test_loader)
  return test_loss
```

```python
class Net(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, layers):
        super(Net, self).__init__()
        self.rnn = torch.nn.LSTM(input_dim, hidden_dim, num_layers=layers, batch_first=True)
        self.fc = torch.nn.Linear(hidden_dim, output_dim, bias=True)

    def forward(self, x):
        x, _status = self.rnn(x)
        x = self.fc(x[:, -1])
        return x

model = Net(input_dim = data_dim, hidden_dim = hidden_dim, output_dim =  output_dim, layers = 1).to(DEVICE)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
train_dataset = CustomDataset(train_x, train_y)
test_dataset = CustomDataset(test_x, test_y)

train_loader = DataLoader(
  train_dataset,
  batch_size = BATCH_SIZE,
  shuffle = True,
  drop_last = False)

test_loader = DataLoader(
  test_dataset,
  batch_size = BATCH_SIZE,
  shuffle = False)

best_val_loss = 100
result_list = []
for epoch in range(1, EPOCHS + 1):
  train_loss = train(model, train_loader)
  test_loss = evaluate(model, test_loader)
  if epoch % 30 == 0:
    print(f"[EPOCH: {epoch}], \tTrain Loss: {train_loss:.4f}, \tTest Loss: {test_loss:.4f}")

  if test_loss <= best_val_loss:

    best_info = {
    'epoch': epoch,
    'model': deepcopy(model.state_dict()),
    'train_loss': train_loss,
    'val_loss': test_loss,
    }

    best_val_loss = test_loss

  result = {
    'EPOCH': epoch,
    'Train Loss': train_loss,
    'Test Loss': test_loss}
  
  result_list.append(result)
```

```python
result_df = pd.DataFrame(result_list)
plt.title('Learning Curves')
plt.plot(result_df['EPOCH'], result_df['Train Loss'], label='Train Loss')
plt.plot(result_df['EPOCH'], result_df['Test Loss'], label='Test Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
model.load_state_dict(best_info['model'])
testX_tensor = torch.Tensor(test_x).to(DEVICE)

plt.plot(test_y)
plt.plot(model(testX_tensor).cpu().data.numpy())
plt.legend(['original', 'prediction'])
plt.show()
```
