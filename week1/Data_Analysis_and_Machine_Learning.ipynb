# 데이터 분석 이란?

세상은 매우 빠른 속도로 발전하고 있다. 기술의 파괴적 혁신 등에 의하여 언제 어떤 기술이 도태되며 개발되며 새로운 시장이 생기고 파괴될지 누구도 쉽게 예상할 수 없다. 세상이 발전 하는 속도 만큼이나 미래에 대한 불확실성 또한 매우 높아지며 미래에 대한 예측이 점점 더 어려줘 지고 있다. 이러한 불확실성이 매우 높은 환경에서 기업은 생존하기 위해서, 성장하기 위해서, 책임을 지기 위해서 효과적인 의사결정을 내려야 하다. 잘못된 의사결정은 곧 기업의 생존에 큰 위협으로 다가 올 것이다.

그렇다면 우리는 효과적인 의사결정을 내리기 위해서 어떻게 해야할 까? 많은 정보를 가지고 있어야 한다. 그러면 정보는 어디에서 얻을 수 있을까? 바로 데이터에서 얻을 수 있다. 따라서 곧 데이터를 분석한다는 것은 정보를 얻는 다는 것이고 정보를 얻는 다는 것은 효과적인 의사결정을 내릴 수 있게 된다는 것이다. 기업은 생존하기 위해서 데이터를 수집하고 분석하며 활용해야 한다.

스마트폰의 출현과 ICT의 발전은 오늘날 우리가 행동하는 모든 것을 데이터로 저장할 수 있게 되었다. 기업은 이제 이러한 데이터를 분석하여 새로운 서비스, 새로운 가치를 창출하고 있다. 그런데 "garbage in, garbage out"이라는 말이 있듯이 데이터를 올바르게 수집하고 분석하는 것이 중요하다. 올바르게 수집되지 못하고 올바르게 분석되지 못한 데이터는 활용할 수 없으며 새로운 가치를 창출할 수 없다.

그렇다면 데이터를 제대로 활용하기 위해서는 어떻게 해야하는 가? 현재 많은 데이터 사이언티스트들이 Python을 활용한다. Python에는 데이터 수집을 위한 BeautifulSoup, 데이터 분석 및 정제를 위한 Numpy, Pandas, seaborn, matplotlib, Scipy 등의 다양한 라이브러리를 무료로 활용할 수 있으며 그 활용성 또한 무궁무진하다. 이러한 오픈소스들을 활용한다면 누구나 쉽게 데이터를 분석하며 활용할 수 있으며 새로운 가치를 창출 할 수 있다.

따라서 앞으로 효과적인 의사결정을 내리기 위해서 꼭 필요한 데이터 분석을 학습하고자 Python을 활용하여 이번 과제를 진행할 예정이다.

# 머신러닝 이란?

Machine Learning, 즉 기계 학습은 데이터를 기반으로 패턴을 학습하고 결과를 예측하는 알고리즘 기법을 칭한다. 따라서 머신러닝과 데이터는 떼려야 뗄 수 없는 서로가 서로에게 영향을 주고 받는 상호보완적인 관계이다.

우리가 데이터 분석을 하는 이유는 효과적인 의사결정을 내리기 위해서라고 아까 이야기 했었다. 그렇다면 데이터를 분석함으로써 효과적인 의사결정을 바로 내릴 수 있을까? 이에 대한 해답은 데이터에 따라 다르겠지만 데이터를 기반으로 예측을 하기 위해서는 데이터 분석만으로는 부족하다. 데이터를 분석해서 얻은 결론을 바탕으로 수식을 세워야지 앞으로를 예측할 수 있다. 예를 들어 $ y = 2x $ 라는 관계를 상품이 있다고 가정할 때 데이터 분석을 통해서 상관계수를 구한다면 쉽게 $ y = 2x $ 라는 수식을 유추할 수 있을 것이다. 그렇지만 실제 세상에는 x라는 feature가 1개가 존재하지 않는다. 10개, 아니 그 이상 수 많은 feature가 존재할 것이며 그 데이터의 양도 매우 많을 것이다. 따라서 사람의 힘으로는 예측을 하는 수식을 만들기 매우 어려울 것이며 시간과 비용아 매우 많이 소모될 것이다. 따라서 우리는 분석된 결과를 바탕으로 머신러닝이 스스로 데이터를 통해서 패턴을 학습해서 결과를 예측 하거나 추천을 하는 애플리케이션을 만들 수 있다.

따라서 데이터 분석에 머신러닝을 활용하다면 사람이 만들기 어려운 수식을 쉽게 만들수 있으며 만드는데 필요한 수많은 비용과 시간을 크게 단축할 수 있어 매우 효율적이며 효과적인 의사결정을 쉽고 빠르게 내릴 수 있다. 빠른 만큼 의사결정을 수정하거나 피드백하는 시간도 단축되어 더 많은 실험을 통해서 최선의 결정을 내릴 수 있을 것이다.

Machine Learning은 크게 
- Supervised Learning (지도학습) : Label 데이터가 있는 상태에서의 학습, 즉 정답을 바탕으로 최대한 정답을 맞출 수 있도록 학습을 하는 방식이다.
 - 분류 : 이산적인 데이터를 맞춘다. ex) 스팸메일 분류
 - 회귀 : 연속적인 데이터를 맞춘다. ex) 자전거 수요량 예측
 - ...
- Unsupervised Learning (비지도 학습) : Label 데이터가 없는 상태에서의 학습, 모델 스스로가 패턴을 찾아서 분류를 하는 학습 방식이다.
 - 군집화 : 모델 스스로가 Label을 매긴다. ex) 고객세분화
 - 차원 축소 : 다차원의 데이터를 저차원의 데이터로 축소하여 시각화를 용이하게 하거나 sparse한 데이터를 일반화가 잘 되는 데이터로 변형시켜주거나 차원을 축소하여 모델의 학습속도를 빠르게 할 수 있다.
 - ...

등으로 나눌 수 있다.

앞으로 효과적인 의사결정을 내리기 위해 데이터를 분석하고 분석된 데이터를 바탕으로 모델 스스로가 데이터를 바탕으로 패턴을 학습하여 예측을 하는 다양한 알고리즘에 대하여 학습할 것이다.

# 데이터 분석의 전체적인 프로세스

데이터 분석의 전체적인 프로세스는 기본적으로

문제 정의 -> 데이터 수집 -> 데이터 전처리 -> Exploratory Data Analysis(EDA) -> Feature Engineering -> Model 설계 및 튜닝 -> 평가 -> 출시 -> 유지 및 보수

의 순으로 진행된다. 

하지만 위의 순서는 솔직히 큰 의미가 없다. 왜냐하면 EDA를 하다가 또는 Model을 설계하다가 언제든 데이터가 부족하다면 다시 데이터 수집 단계로 넘어갈 수도 있으며 프로세스를 진행하다가 새로운 문제를 정의할 수도 있으며 EDA를 통해서 결측치를 발견하여 다시 전처리를 하거나 등의 일이 계속해서 반복되기 때문이다. 또한 데이터 분석은 하나의 프로세스가 끝났다고 해서 끝이 아니다. 계속해서 반복을 하는 Cycle 형태이다. 예를 들어 하나의 추천시스템을 만들었다고 가정했을 때 추천시스템을 만들었다고 해서 끝이 아니며 그 추천시스템이 만들어지는 동안 또는 출시된 동안 또 다시 새롭게 데이터가 축척되었을 것이며 축척된 데이터를 바탕으로 모델을 다시 유지 및 보수하며 업데이트를 하는 과정을 거쳐야 하며 새로운 문제를 정의하여 다시 새로운 Task를 실행할 수도 있을 것이다.

따라서 데이터 분석은 전체적인 프로세스는 따르며 그 프로세스의 단계에 얽매이지 않고 언제든 필요한 프로세스 단계로 진입하여 문제를 해결해야 할 것이다.

# 데이터 수집

우리는 데이터 수집을 할 때 머릿속에 새겨야 할 말이 있다. 바로 "garbage in, garbage out" 이다. 쓰레기를 넣으면 쓰레기가 나온다. 즉 쓰레기 데이터를 활용한다면 결과물 또한 쓰레기가 나올 것이며 활용할 수 없을 것이고 모델을 만드는데 소요된 시간과 비용이 모두 헛투로 사용되었으므로 큰 손해가 될 것이다. 따라서 우리가 정의 문제에 맞는 제대로 된 데이터를 수집하는 것은 매우 중요하다.

여기에서는 데이터를 수집하는 가장 기초적이면서 필수적인 방법인 Crawling에 대하여 Python을 활용하여 학습하겠다.

BeautifulSoup 라이브러리를 활용하여 https://finance.naver.com/sise/sise_market_sum.nhn?&page=1 에 있는 국내 코스피 기업을 모두 수집할 예정이다.

스크립트 기반의 크롤링

```
# 사용할 라이브러리 불러오기
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```
# html에서 title, url을 수집하기 위한 함수
def get_finance_list(html):  
  soup = BeautifulSoup(html)
  
  elements = soup.select('.box_type_l tbody tr')
  
  finance_list = []
  for element in elements:
    link_element = element.select_one('a')
    result = dict()
    if link_element:
      result['title'] = link_element.text
      result['url'] = 'https://finance.naver.com' + link_element['href']
      finance_list.append(result)
      
  return finance_list

# 전체 기업을 수집하기 위한 함수
def get_all_finance_list():
  # 크롤링할 페이지가 저장될 리스트
  all_finance_list = []

  # 초기값 설정
  count = 1
  finance_list = [1]

  # 크롤링 할 페이지기 없다면 반복을 중단
  while finance_list:
    finance_list_url = "https://finance.naver.com/sise/sise_market_sum.nhn?&page="+ str(count)
    response = requests.get(finance_list_url)
    finance_list = get_finance_list(response.content)
    
    all_finance_list += finance_list
    count +=1
      
  return all_finance_list
```

```
# 국내증시 시가총액 페이지에서 기업 리스트를 읽어옵니다.
all_finanace_list = get_all_finance_list()

# 기업 리스트를 데이터 프레임 형태로 저장
all_finanace_list_df = pd.DataFrame(all_finanace_list)

# 데이터를 6개만 출력
all_finanace_list_df.head()
```

```
print(f'총 수집된 기업의 수: {len(all_finanace_list_df)}개')
```

위와 같이 크롤링을 통하여 우리는 원하는 정보를 수잡할 수 있다. 하지만 크롤링을 통하여 수집할 수 없는 정보도 많으며 대부분의 정보가 크롤링을 통해서 수집할 수 없을 것이다. 이럴때 우리는 공공데이터 포털 또는 기업이라면 자신의 데이터 베이스를 활용하여 목적에 적합한 데이터를 수집해야할 것이다. 

# 데이터 전처리

데이터 전처리란 분석에 적합한 형태로 데이터를 가공하는 작업을 의미한다. 실제 현업에서 활용되는 데이터는 분석 또는 머신러닝에 활용하기에 적합하지 않은 경우가 많으며, NA라고 하는 결측치도 많이 존재한다. 이러한 데이터를 분석하기 용이하게 처리하는 단계가 데이터 전처리이며 분석 결과를 바탕으로 결측치를 채우는 것도 데이터 전처리 이다. 

그런데 데이터를 분석하기 용이하게 처리하기 위해서, 데이터의 결측치를 채우기 위해서는 데이터를 탐색해야 하며 feature를 변경해야 한다. 이렇듯 데이터 전처리 내에서도 EDA, Feature Engineering 등의 수행되기 때문에 사실상 데이터 전처리, EDA, Feature Engineering 간의 경게는 모호하며 현재의 상황에 알맞은 작업을 분석가가 올바르게 수행해야 한다.

아래의 코드는 아까 수집한 코스피에 존재 하는 국내 기업의 리스트에 종목코드를 얻는 과정이다.

```
all_finanace_list_df.head()
```

url 컬럼을 보면 code= 뒤의 숫자가 해당 기업의 주식 코드라는 것을 알 수 있다. 따라서 해당 기업의 주식 코드를 얻기 위해서 데이터를 전처리 한후 새롭게 code라는 컬럼을 생성하겠다.

```
# 정규표현식을 활용하여 문자열을 쉽게 처리하기 위한 라이브러리
import re

all_finanace_list_df['code'] = all_finanace_list_df['url'].apply(lambda x : re.sub('[^0-9]', "", x))
all_finanace_list_df.head()
```

위와 같이 새옵게 code라는 컬럼이 생성된 것을 확인 할 수 있다. 우리는 이제 unique한 code를 활용하여 주식 사이트에서 새로운 데이터를 수집할 수도 있을 것이다.

# EDA

Exploratory Data Analysis는 탐색적 데이터 분석(기존에 데이터에 대한 경험이 없을 때 하는 분석)이라고 부르며 한마디로 요약하자면 내가 탐정이 되어서 범인을 찾는 과정이라고 할 수 있다. 탐정이 범인을 찾을 때 모든 경우의 수를 확인하고 모르는 것이 없을 때까지 탐색하듯이 EDA에서는 내가 탐정이 되어서 범인을 찾듯이 데이터에 대해서 내가 모르는 것이 없을 정도로 탐색해야 한다.

기본적으로 데이터의 분포, 상관 관계, 인과 관계 등을 탐색해야 하며 탐색하는 방식은 그래프(boxplot, countplot, distplot 등) 등을 활용하여 알 수도 있고 또는 직접 수치(상관 계수, max, min, mean, IQR 등)를 통해서도 확인할 수 있다.

EDA에는 정답이란 없으며 가장 중요한 것은 내가 데이터에 대해서 모르는 것이 없을 때 까지 분석을 진행해야 한다는 것이다.

아래는 Iris 데이터의 분포와 기본적인 통계치를 확인한 EDA의 기본 예시이다.

```
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

# 경고무시
import warnings
warnings.filterwarnings(action='ignore') 

iris = load_iris()
iris_data = iris.data

df = pd.DataFrame(data = iris_data, columns = iris.feature_names)
df.head()
```

```
# 기본적인 통계치 확인
df.describe()
```

```
# 컬럼별 데이터의 분포 확인
cols = df.columns
figure, ax = plt.subplots(nrows=len(cols), ncols=1)

figure.set_size_inches(18,15)
for idx, col in enumerate(cols):
  sns.distplot(df[col], ax= ax[idx])
```

# Feature Engineering

Feature Engineering은 모델의 예측력을 높이기 위한 변수를 새롭게 만들거나 변수를 모델에 넣기 위하여 처리하는 모든 과정을 포함한다.

모델의 예측력을 높이기 위해서 새롭게 변수를 만드는 과정은 기본적으로 EDA가 필요하기 때문에 추후에 실습에서 다룰 예정이며 EDA가 없어도 기본적으로 알아야 할 Feature Engineering 기법이 있다.

바로 Scaling과 Encoding 이다.

- Scaling : 데이터의 분포를 상황에 알맞게 조정하는 방법이다.
 - StandardScaler : 표준화로 평균을0, 분산을 1로 변경해주며 모든 픽쳐들이 동일한 스케일을 갖게 된다.
 - RobustScaler : 표준화에서 평균과 분산대신 중앙값과 분위수를 활용하는 방법으로 평균을 사용하지 않고 중앙값을 사용하기 때문에 이상치에 영향을 받지 않는다는 특징을 가지고 있다.
 - MinMaxScaler : 모드 feature를 0과 1사이에 위치하게 만든다.
 - Log transformation : 매우 큰 값을 상대적으로 줄여주며 한쪽으로 기울어진 분포를 정규분포와 유사하게 만들어주는 특징을 가진다.
 - Centering : 분산을 건드리지 않는 상태에서 데이터의 평균을 0으로만 만드는 방식이다. 다중공선성을 줄이기 위해서 사용한다.
 - .....
- Encoding : 범주형 변수를 모델이 이해할 수 있게 숫자로 변환하는 과정이다.
 - One-Hot-Encoding : 하나의 컬럼에 있는 모든 카테고리 값을 컬럼으로 활용하는 방식이다. 차원이 늘어난다는 특징을 가지며 선형회귀의 경우 활용성이 높아진다.
 - Label Encoding : 하나의 컬럼에 있는 모든 카테고리를 숫자로 변환하는 방식이다. 차원이 늘어나지 않는 특징을 가지며 선형 회귀에는 사용할 수 없으며 카테고리 자체에 순서가 생길수도 있다는 위험을 가진다.
 - ....

우리가 이러한 Scaling과 Encoding에 대하여 자세히 알아야 하는 이유는 사용할 알고리즘에 맞게 데이터의 형태로 변형 시킬 수 있어야지 머신러닝 모델의 성능을 더 증가 시킬 수 있기 때문이다.

```
# 표준화 예시
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(df)

standard_df = scaler.transform(df)

standard_df = pd.DataFrame(data = standard_df, columns = iris.feature_names)

cols = standard_df.columns
figure, ax = plt.subplots(nrows=len(cols), ncols=1)

figure.set_size_inches(18,15)
for idx, col in enumerate(cols):
  sns.distplot(standard_df[col], ax= ax[idx])
```

```
# 정규화 예시
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df)
minmax_df = scaler.transform(df)

minmax_df = pd.DataFrame(data = minmax_df, columns = iris.feature_names)

cols = minmax_df.columns
figure, ax = plt.subplots(nrows=len(cols), ncols=1)

figure.set_size_inches(18,15)
for idx, col in enumerate(cols):
  sns.distplot(minmax_df[col], ax= ax[idx])
```

```
# 로그변환 예시
log_df = np.log(df)

cols = log_df.columns
figure, ax = plt.subplots(nrows=len(cols), ncols=1)

figure.set_size_inches(18,15)
for idx, col in enumerate(cols):
  sns.distplot(log_df[col], ax= ax[idx])
```

# Model 설계 및 튜닝

기본적으로 우리가 모델을 만드는 이유는 수집하지 못한 즉 앞으로의 데이터에 대하여 예측을 하기 위해서 이다. 따라서 기본적으로 모델의 성능을 평가하기 위해서 Train, Validation, Test 용 데이터로 분리를 해야 한다. 우리는 Test 데이터가 앞으로 우리가 맞딱드릴 데이터라는 가정하에 Train과 Validation 데이터를 활용하여 모델을 설계하고 튜닝해야 한다.

모델은 최대한 다양하고 많은 모델을 활용해야 한다. 왜냐하면 NFL(No Free Lunch) 이론에 따르면 어떤 데이터가 어떠한 모델에 적합할지 그 누구도 알 수 없기 때문에 최대한 다양한 모델을 활용하여 우리의 데이터 셋에 적합한 모델을 찾아야 한다.

적합한 모델을 찾은 후에는 이제 모델의 성능을 높이기 위해서 하이퍼 파라미터 튜닝을 통해서 최적의 파라미터를 찾아 모델의 성능을 높여야 한다.

하이퍼 파라미터 튜닝의 방식으로는 크게

- Random Search : 파라미터의 특정한 범위를 정해서 랜덤으로 파라미터를 찾는 방식이다. 특정한 데이터의 경우 Grid Search 보다 시간이 적게 걸릴 수 있으며 모든 파라미터를 설정할 수 있다는 장점이 있다.
- Grid Search : 파라미터를 직접 지정하여 최적의 파라미터를 찾는 방식이다. Random Search 보다 시간이 적게 걸릴 가능성은 높지만 내가 지정하지 않은 파라미터 내에서 최적의 모델의 성능이 생길 가능성이 존재한다.

로 나뉘어진다.



```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
 
# 데이터셋 라이브러리 로드
from sklearn.datasets import load_iris

import warnings
warnings.filterwarnings('ignore')
```

```
# 데이터셋 로드
iris = load_iris()

# target
iris_label = iris.target

# features
iris_features = iris.data
```

분류 데이터의 경우 기본적으로 stratify하게 분리를 해야한다.
왜냐하면 데이터의 분포가 달라진다면 그만큼 성능이 달라지며 모델의 성능을 제대로 평가할 수 없게 되기 때문이다.

```
# Random한 데이터 분리
from sklearn.model_selection import train_test_split
import seaborn as sns
X_train, X_test, y_train, y_test = train_test_split(iris_features, iris_label, random_state=22, test_size=0.2)

figure, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
sns.countplot(y_train, ax = ax[0])
ax[0].set_title('Train Random')

sns.countplot(y_test, ax = ax[1])
ax[1].set_title('Test Random')
```

위와 같이 Random 하게 데이터를 분리할 경우 Train 데이터의 경우 Test 데이터와 달리 2번 라벨의 수가 적어 2번 라벨을 제대로 못 맞출 가능성이 높아져 모델의 성능을 제대로 평가할 수 없을 것이다.

```
# stratify한 데이터 분리
from sklearn.model_selection import train_test_split
import seaborn as sns
X_train, X_test, y_train, y_test = train_test_split(iris_features, iris_label, random_state=22, test_size=0.2, stratify = iris_label)

figure, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
sns.countplot(y_train, ax = ax[0])
ax[0].set_title('Train Stratify')

sns.countplot(y_test, ax = ax[1])
ax[1].set_title('Test Stratify')
```

위와 같이 대이터를 stratify하게 나누면 데이터의 라벨이 동등하게 나뉘어 진다는 것을 확인할 수 있다.

```
# RandomForestClassifier 모델을 활용하여 oof 방식의 모델 성능 평가
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from tqdm import tqdm_notebook

rf = RandomForestClassifier(random_state = 22)

rf_oof = np.zeros(y_train.shape)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=22)

for idx, (trn_idx, val_idx) in tqdm_notebook(enumerate(skf.split(X_train, y_train))):
  x_trn, x_val = X_train[trn_idx], X_train[val_idx]
  y_trn, y_val = y_train[trn_idx], y_train[val_idx]

  rf.fit(x_trn, y_trn)
  pred = rf.predict(x_val)

  # 모델 정확도 측정
  print("{0}번째 RandomForestClassifier의 정확도 : {1:.3f}, 학습 데이터의 크기: {2}, 검증데이터의 크기: {3}".format(idx+1 ,accuracy_score(y_val, pred), x_trn.shape[0], x_val.shape[0]))

  rf_oof[val_idx] = pred

# 모델 정확도 측정
print("RandomForestClassifier의 정확도 : {0:.4f}".format(accuracy_score(y_train, rf_oof)))
```

```
# GridSearch를 통한 최적의 파라미터 찾기
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state = 22)

params = {'max_depth':[2, 3, 4, 5, 6, 7, 8],
          'min_samples_split':[2, 3, 4, 5, 6]}

grid_rf = GridSearchCV(rf, param_grid=params, cv=5, refit=True)

grid_rf.fit(X_train, y_train)

print('최적의 파라미터: ', grid_rf.best_params_)
print('최고의 정확도: ', grid_rf.best_score_)
```

# 평가

Test 데이터에 대한 모델의 성능을 평가하여 현재 모델을 적용할 것인지를 결정한다.

```
from sklearn.metrics import accuracy_score

# 최적의 파라미터로 학습
estimator = grid_rf.best_estimator_

pred = estimator.predict(X_test)
print('정확도: ', accuracy_score(y_test, pred))
```
