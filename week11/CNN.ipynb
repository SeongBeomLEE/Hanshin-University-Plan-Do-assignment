# CNN 이란?

Convolutional Neural Network(CNN), 합성곱 신경망은 주로 이미지, 컴퓨터 비전 분야에서 사용되는 MLP의 한 종류이다.

앞서 우리가 배운 MLP의 경우에도 이미지 데이터에 대하여 사용할 수 있었지만 CNN은 기본적인 MLP보다 이미지 데이터에 대하여 월등한 성능을 보여준다. 왜냐하면 CNN은 합성곱이라는 연산 방식을 통하여 이미지의 공간적인 구조 정보를 보존하면서 학습할 수 있기 때문이다. 

CNN은 Filter를 통하여 가중치를 공유하기 때문에 이미지 내에서의 위치와 관계없이 동일한 패턴을 인식하는 위치이동불변의 특성을 가진다. 그리고 Convolutional Layer와 Pooling Layer로 이루어진 CNN은 각 층을 통하여 계층적으로 이미지의 특성을 추출하는 특징을 가진다.

컴퓨터 비전 분야에서 월등한 성능을 보이는 CNN을 이해하기 위해서는 기본적으로 Channel, Filter, Stride, Padding의 개념이 필요하고 이 개념을 가지고 CNN을 구성하는 Convolutional Layer와 Pooling Layer를 설명하도록 하겠다.



## Channel

대부분의 이미지는 RGB 3개의 색깔로 이루어져 있다. 그리고 이미지에는 가로와 세로라는 크기가 주어진다. 이를 텐서로 표현하면 (높이, 너비, 채널)로 표현하게 되는데 여기서 채널이 색깔을 의미한다. 만약 흑백의 가로 세로 28크기의 이미지가 있다면 텐서로 (28, 28, 1)로 표현할 수 있다.

## Filter

Filter(= Kernel)는 합성곱 연산에서 이미지의 특징을 추출하기 위한 필요한 파라미터의 집합이라고 할 수 있다. 아래의 그림 처럼 input이 들어오면 Filter와 합성곱을 통하여 output이 만들어지게 된다. 여기서 Filter을 사용하여 합성곤 연산을 통해 나온 output을 Feature Map 이라고 한다.

image.png

## Stride

Stride는 위에 그림에서 Filter가 한번 연산을 하기 위해 움직이는 거리를 의미한다.

## Padding

Padding은 input과 Feature Map의 크기를 동일하게 유지하기 위해서 사용하는 방법이다. 패딩은 입력의 가장자리에 지정된 개수의 폭만큼 행과 열을 추가해주는 것을 말하며 주로 값을 0으로 채우는 zero padding을 사용한다.

image.png

# Convolutional Layer

Convolutional Layer는 합섭공 연산을 통하여 이미지의 특징을 추출하는 역할을 한다. 위에서 다룬 Filter, Stride, Padding의 개념을 통하여 각 채널마다 Filter를 통해서 Feature Map를 추출하고 추출된 Feature Map은 다시 Convolutional Layer 또는 Pooling Layer를 거쳐서 계속해서 Feature Map을 추출하고 이러한 층을 반복하여 CNN은 Input data를 학습한다. 그리고 Loss를 바탕으로 CNN은 역전파를 통해 Filter의 가중치를 업데이트하고 최적의 가중치가 나올때 까지 계속하여 반복을 진행한다. CNN의 목적은 최적의 Filter를 찾는 것이라고 볼 수 있다.

image.png

# Pooling Layer

Pooling Layer는 이미지의 특성을 유지하면서 사이즈를 줄이는 역할을 한다. 아래의 그림 처럼 학습이 필요 없는 Filter를 지정한 stride 만큼 이동시켜서 최대, 평균, 최소 등의 기준을 통하여 이미지의 특성을 추출하여 이미지의 크기를 줄이는 방식이다. 대부분의 Pooling Layer는 최대값을 출력하는 Max Pooling 방식을 많이 사용한다.

# CNN 아키텍처의 종류

CNN 아키텍쳐의 종류로는 LeNet, AlexNet, ZFNet, VGG, GoogLeNet, ResNet 등이 존재하며 이 밖에도 다양한 아키텍쳐 들이 존재한다. 이 중에서도 ImageNet 데이터셋을 이용하여 Classification 성능을 겨루는 대회인 ILSVRC 대회에서 2012년에 압도적인 성능으로 우승을 하여 화제를 모았던 AlexNet과 CNN의 Layer을 깊게 쌓는 방법을 알아내어 현재 끼지도 대부분의 CNN의 Base 모델로 활용되는 ResNet에 대하여 알아보고자 한다.

- AlexNet: AlexNet은 2012년 NIPS에 발표된 ImageNet Classification with Deep Convolutional Neural Networks 논문에서 소개된 모델로 논문의 저자의 이름인 Alex와 Network를 합쳐서 AlexNet이라고 불리고 있다. AlexNet은 활성화 함수로 ReLU를 활용했다는 것, 훈련시에 2개의 GPU로 나누어 학습을 했다는 것, Local Response Normalization을 활용했다는 것, Overlapping Pooling을 활용했다는 것, Data Augmentation을 활용했다는 것, Dropout을 활용했다는 것을 특징으로 가지고 있다.

- ResNet: ResNet은 2015년 Microsoft Research에서 발표된 Deep Residual Learning for Image Recognition 논문에서 소개된 모델이다. ResNet이라는 이름은 본 몬문에서 제안한 핵심 아이디어인 Residual Block에서 유래 되었다. ResNet의 가장 큰 특징이자 현재 까지도 많이 활용되고 있는 Residual learning은 기존의 input을 어느 정도 유지한채로 학습을 하는 방식으로 input을 Convolutional Layer에 통과 시켜 나온 output에 기존의 input을 더해가면서 계속하여 이미지의 특징을 추출하는 방식이다. 이러한 방식 덕분에 CNN의 층을 더욱더 깊게 쌓을 수 있는 요건을 충족할 수 있게 되었다.

# CNN 실습

아래의 코드는 ResNet을 직접 구현하여 MNIST 데이터를 분류하는 실습을 진행한 코드이다. 학습 결과 Test Data에서 약 99.43%의 성능을 보인다는 것을 알 수 있으며 또한 MLP보다 더 우수한 성능을 보인다는 것도 코드를 통해서 알 수 있다. 이렇듯 CNN은 이미지 데이터에 대하여 우수한 성능을 보이는 MLP의 한 종류이다.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms, datasets
```

```python
if torch.cuda.is_available():
  DEVICE = torch.device('cuda')
else:
  DEVICE = torch.device('cpu')
```

```python
BATCH_SIZE = 128
EPOCHS = 10
```

```python
train_dataset = datasets.MNIST(root = "../data/MNIST",
                               train = True,
                               download = True,
                               transform = transforms.ToTensor())

test_dataset = datasets.MNIST(root = "../data/MNIST",
                               train = False,
                               transform = transforms.ToTensor())

train_loader = torch.utils.data.DataLoader(dataset = train_dataset,
                                           batch_size = BATCH_SIZE,
                                           shuffle = True)

test_loader = torch.utils.data.DataLoader(dataset = test_dataset,
                                           batch_size = BATCH_SIZE,
                                           shuffle = False)
```

```python
for (X_train, y_train) in train_loader:
  break

pltsize = 1
plt.figure(figsize=(10 * pltsize, pltsize))
for i in range(10):
  plt.subplot(1, 10, i + 1)
  plt.axis('off')
  plt.imshow(X_train[i, :, :, :].numpy().reshape(28, 28), cmap='gray_r')
  plt.title('Class: ' + str(y_train[i].item()))
```

```python
def train(model, train_loader):
  model.train()
  train_loss = 0
  correct = 0

  for batch_idx, (image, label) in enumerate(train_loader):
    image = image.to(DEVICE)
    label = label.to(DEVICE)
    optimizer.zero_grad()
    output = model(image)
    loss = criterion(output, label)
    loss.backward()
    optimizer.step()

    train_loss += loss.item()
    prediction = output.max(1, keepdim = True)[1]
    correct += prediction.eq(label.view_as(prediction)).sum().item()
      
  train_loss /= len(train_loader.dataset)
  train_accuracy = 100. * correct / len(train_loader.dataset)
  return train_loss, train_accuracy
```

```python
def evaluate(model, test_loader):
  model.eval()
  test_loss = 0
  correct = 0

  with torch.no_grad():
    for image, label in test_loader:
      image = image.to(DEVICE)
      label = label.to(DEVICE)
      output = model(image)
      test_loss += criterion(output, label).item()
      prediction = output.max(1, keepdim = True)[1]
      correct += prediction.eq(label.view_as(prediction)).sum().item()

  test_loss /= len(test_loader.dataset)
  test_accuracy = 100. * correct / len(test_loader.dataset)
  return test_loss, test_accuracy
```

```python
class BasicBlock(nn.Module):
  def __init__(self, in_planes, planes, stride = 1):
    super(BasicBlock, self).__init__()
    self.conv1 = nn.Conv2d(in_planes, planes,
                           kernel_size = 3,
                           stride = stride,
                           padding = 1,
                           bias = False)
    self.bn1 = nn.BatchNorm2d(planes)
    self.conv2 = nn.Conv2d(planes, planes,
                           kernel_size = 3,
                           stride = 1,
                           padding = 1,
                           bias = False)
    self.bn2 = nn.BatchNorm2d(planes)

    self.shortcut = nn.Sequential()
    if stride != 1 or in_planes != planes:
      self.shortcut = nn.Sequential(
          nn.Conv2d(in_planes, planes,
                           kernel_size = 3,
                           stride = stride,
                           padding = 1,
                           bias = False),
          nn.BatchNorm2d(planes))
      
  def forward(self, x):
    out = F.relu(self.bn1(self.conv1(x)))
    out = self.bn2(self.conv2(out))
    out += self.shortcut(x)
    out = F.relu(out)
    return out

class ResNet(nn.Module):
  def __init__(self, block, num_blocks, num_classes = 10):
    super(ResNet, self).__init__()
    self.in_planes = 16
    
    self.conv1 = nn.Conv2d(1, 16,
                           kernel_size = 3,
                           stride = 1,
                           padding = 1,
                           bias = False)
    self.bn1 = nn.BatchNorm2d(16)
    self.layer1 = self._make_layer(block, 16, num_blocks[0], stride = 1)
    self.layer2 = self._make_layer(block, 32, num_blocks[1], stride = 2)
    self.layer3 = self._make_layer(block, 64, num_blocks[2], stride = 2)
    self.layer4 = self._make_layer(block, 128, num_blocks[3], stride = 2)
    self.linear = nn.Linear(128, num_classes)

  def _make_layer(self, block, planes, num_blocks, stride):
    strides = [stride] + [1] * (num_blocks - 1)
    layers = []
    for stride in strides:
      layers.append(block(self.in_planes, planes, stride))
      self.in_planes = planes
    return nn.Sequential(*layers)

  def forward(self, x):
    out = F.relu(self.bn1(self.conv1(x)))
    out = self.layer1(out)
    out = self.layer2(out)
    out = self.layer3(out)
    out = self.layer4(out)
    out = F.avg_pool2d(out, 4)
    out = out.view(out.size(0), -1)
    out = self.linear(out)

    return out
```

```python
model = ResNet(BasicBlock, [2, 2, 2, 2]).to(DEVICE)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
criterion = nn.CrossEntropyLoss()

print(model)
```

```python
result_list = []
for epoch in range(1, EPOCHS + 1):
  train_loss, train_accuracy = train(model, train_loader)
  test_loss, test_accuracy = evaluate(model, test_loader)
  print(f"\n[EPOCH: {epoch}], \tTrain Loss: {train_loss:.4f}, \tTrain Accuracy: {train_accuracy:.2f} %, \tTest Loss: {test_loss:.4f}, \tTest Accuracy: {test_accuracy:.2f} % \n")
  result = {
    'EPOCH': epoch,
    'Train Loss': train_loss,
    'Train Accuracy': train_accuracy,
    'Test Loss': test_loss,
    'Test Accuracy': test_accuracy}
  
  result_list.append(result)
result_df = pd.DataFrame(result_list)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
fig.suptitle('Learning Curves')
axes[0].plot(result_df['EPOCH'], result_df['Train Loss'], label='Train Loss')
axes[0].plot(result_df['EPOCH'], result_df['Test Loss'], label='Test Loss')
axes[0].set_ylim(0.0, 0.01)
axes[0].legend()

axes[1].plot(result_df['EPOCH'], result_df['Train Accuracy'], label='Train Accuracy')
axes[1].plot(result_df['EPOCH'], result_df['Test Accuracy'], label='Test Accuracy')
axes[1].set_ylim(80, 100)
axes[1].legend()
plt.tight_layout()
plt.show()
```
